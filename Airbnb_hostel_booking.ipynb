```python

# Remove duplicates
df.drop_duplicates(inplace=True)

# Clean price and service fee columns (with proper error handling)
# Common column name variations - adjust based on your actual column names

# Option 1: If column names are exactly 'price' and 'service fee'
try:
    df['price'] = df['price'].str.replace('$', '', regex=False)
    df['service fee'] = df['service fee'].str.replace('$', '', regex=False)
    df['price'] = df['price'].str.replace(',', '', regex=False)  # Fixed: removed space after comma
    df['service fee'] = df['service fee'].str.replace(',', '', regex=False)  # Fixed: removed space after comma
except KeyError as e:
    print(f"Column not found: {e}")
    # Find similar column names
    price_cols = [col for col in df.columns if 'price' in col.lower()]
    service_cols = [col for col in df.columns if 'service' in col.lower() and 'fee' in col.lower()]
    print(f"Similar price columns: {price_cols}")
    print(f"Similar service fee columns: {service_cols}")

# Fix the rename syntax (original had syntax errors)
df.rename(columns={
    'price': 'price_$',
    'service fee': 'service_fee_$'
}, inplace=True)

# Remove missing values
df.dropna(inplace=True)

# Convert data types with error handling
try:
    df['price_$'] = df['price_$'].astype(float)
    df['service_fee_$'] = df['service_fee_$'].astype(float)
    df['id'] = df['id'].astype(str)
    df['host id'] = df['host id'].astype(str)
    df['last review'] = pd.to_datetime(df['last review'])
    df['Construction year'] = df['Construction year'].astype(int)
except KeyError as e:
    print(f"Error converting column: {e}")

# Fix neighborhood group typo
if 'neighbourhood group' in df.columns:
    df.loc[df['neighbourhood group'] == 'brookln', 'neighbourhood group'] = 'Brooklyn'

# Remove outliers in availability
if 'availability 365' in df.columns:
    df = df.drop(df[df['availability 365'] > 500].index)

print("\nData cleaning completed successfully!")
print("Final DataFrame shape:", df.shape)
print("\nData types after cleaning:")
print(df.dtypes)
```
